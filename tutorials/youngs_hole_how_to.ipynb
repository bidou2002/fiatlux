## Three holes Young's experiment

```python
# DEFINE FIELD SIZE
n = 256

#  CREATE A FLAT TITLTED FIELD WITH ARBITRARY TIP AND TILT
left_hole = Field(n,
                  field_map='gaussian',
                  gaussian_center=[-10,0],
                  gaussian_variance = 1,
                  )
right_hole = Field(n,
                   field_map='gaussian',
                   gaussian_center=[10,0],
                   gaussian_variance = 1,
                   )
top_hole = Field(n,
                field_map='gaussian',
                gaussian_center=[0,50],
                gaussian_variance = 1,
                )

# CREATE A DETECTOR
cam = Detector(256,
               display_intensity=True,
               )

FFT = Propagator('FFT', left_hole)
IFFT = Propagator('IFFT', left_hole)

((left_hole * IFFT) + (right_hole * IFFT) + (top_hole * IFFT) ).disp_complex_amplitude()
```
